```python
import matplotlib.pyplot as plt


makes = ["Toyota", "Honda", "Toyota", "BMW", "Nissan", "Toyota", "Honda", "Honda", "Toyota", "Nissan"]


make_counts = {}
for make in makes:
    make_counts[make] = make_counts.get(make, 0) + 1


plt.bar(make_counts.keys(), make_counts.values())
plt.title("Number of Cars per Make (Matplotlib Only)")
plt.xlabel("Car Make")
plt.ylabel("Count")
plt.show()
```

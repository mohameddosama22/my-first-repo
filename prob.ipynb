```python


import numpy as np

ages = [18, 21, 20, 19, 22, 23, 20, 21, 22, 19]
mean_manual = sum(ages) / len(ages)
mean_numpy = np.mean(ages)

print("1️ Mean Calculation")
print("Manual Mean:", mean_manual)
print("NumPy Mean:", mean_numpy)
print("-" * 50)

random_list = np.random.randint(10, 51, size=15)
sorted_list = sorted(random_list)
n = len(sorted_list)
if n % 2 == 1:
    median_manual = sorted_list[n // 2]
else:
    median_manual = (sorted_list[n // 2 - 1] + sorted_list[n // 2]) / 2
median_numpy = np.median(random_list)

print("2️ Median Without Sorting")
print("Random List:", random_list)
print("Manual Median:", median_manual)
print("NumPy Median:", median_numpy)
print("-" * 50)

min_manual = min(random_list)
max_manual = max(random_list)
range_manual = max_manual - min_manual
min_numpy = np.min(random_list)
max_numpy = np.max(random_list)
range_numpy = np.ptp(random_list)

print("3️ Range and Data Spread")
print(f"Min: {min_manual}, Max: {max_manual}")
print("Manual Range:", range_manual)
print("NumPy Range:", range_numpy)
print("-" * 50)

data = [5, 7, 3, 7, 9, 10, 15, 7]
mean_data = sum(data) / len(data)
variance_manual = sum((x - mean_data) ** 2 for x in data) / len(data)
std_manual = variance_manual ** 0.5
variance_numpy = np.var(data)
std_numpy = np.std(data)

print("4️ Variance & Standard Deviation")
print("Variance (Manual):", variance_manual)
print("Variance (NumPy):", variance_numpy)
print("Std Dev (Manual):", std_manual)
print("Std Dev (NumPy):", std_numpy)
print("-" * 50)


q25, q50, q75 = np.percentile(data, [25, 50, 75])
iqr = q75 - q25

print("5️ Quartiles & IQR")
print("Q1:", q25)
print("Q2 (Median):", q50)
print("Q3:", q75)
print("IQR:", iqr)
print("-" * 50)

def z_score_normalization(lst):
    mean_lst = sum(lst) / len(lst)
    std_lst = (sum((x - mean_lst) ** 2 for x in lst) / len(lst)) ** 0.5
    return [(x - mean_lst) / std_lst for x in lst]

z_scores_manual = z_score_normalization(data)
z_scores_numpy = (np.array(data) - np.mean(data)) / np.std(data)

print("6️ Z-Score Normalization")
print("Manual Z-Scores:", z_scores_manual)
print("NumPy Z-Scores:", z_scores_numpy)
print("-" * 50)

random_20 = np.random.randint(1, 101, size=20)
window_size = 3
moving_avg_manual = [np.mean(random_20[i:i+window_size]) for i in range(len(random_20) - window_size + 1)]
moving_avg_numpy = np.convolve(random_20, np.ones(window_size)/window_size, mode='valid')

print("7️ Moving Average (Window=3)")
print("Random List (20 numbers):", random_20)
print("Manual Moving Average:", moving_avg_manual)
print("NumPy Moving Average:", moving_avg_numpy)
print("-" * 50)

numbers = list(np.random.randint(1, 51, size=25)) + [300, 400]
q1, q3 = np.percentile(numbers, [25, 75])
iqr_numbers = q3 - q1
lower_bound = q1 - 1.5 * iqr_numbers
upper_bound = q3 + 1.5 * iqr_numbers
outliers = [x for x in numbers if x < lower_bound or x > upper_bound]

print("8️ Outlier Detection (IQR Method)")
print("Numbers:", numbers)
print("Outliers:", outliers)
```
